Thu thập dữ liệu

```python
import pandas as pd

#Đọc dữ liệu 
df = pd.read_csv("calories.csv")

#Hiển thị 5 dòng đầu tiên
df.head()
```

Hiển thị thông tin tổng quan của bộ dữ liệu

```python
# Kiểm tra kiểu dữ liệu, số lượng giá trị thiếu
df.info()

# Thống kê dữ liệu (trung bình, min, max...)
print(df.describe().to_string())
```

Phân tích và trực quan hóa dữ liệu

Trực quan hóa dữ liệu

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Biểu đồ đếm giới tính
plt.figure(figsize=(8, 4))
sns.countplot(x="Gender", data=df, palette="pastel")
plt.title("Phân bố giới tính")
plt.show()

# Biểu đồ phân phối tuổi
plt.figure(figsize=(8, 4))
sns.histplot(df["Age"], kde=True, color="salmon", bins=30)
plt.title("Phân phối tuổi")
plt.xlabel("Tuổi")
plt.ylabel("Tần suất")
plt.show()

# Biểu đồ phân phối chiều cao
plt.figure(figsize=(8, 4))
sns.histplot(df["Height"], kde=True, color="salmon", bins=40)
plt.title("Phân phối chiều cao")
plt.xlabel("Chiều cao (cm)")
plt.ylabel("Tần suất")
plt.show()

# Biểu đồ phân phối cân nặng
plt.figure(figsize=(8, 4))
sns.histplot(df["Weight"], kde=True, color="salmon", bins=40)
plt.title("Phân phối cân nặng")
plt.xlabel("Cân nặng (kg)")
plt.ylabel("Tần suất")
plt.show()

# Biểu đồ phân phối thời gian tập luyện
plt.figure(figsize=(8, 4))
sns.histplot(df["Duration"], kde=True, color="salmon", bins=30)
plt.title("Phân phối thời gian tập luyện")
plt.xlabel("Thời gian tập luyện (phút)")
plt.ylabel("Tần suất")
plt.show()

# Biểu đồ phân phối nhịp tim
plt.figure(figsize=(8, 4))
sns.histplot(df["Heart_Rate"], kde=True, color="salmon", bins=30)
plt.title("Phân phối nhịp tim")
plt.xlabel("Nhịp tim (bpm)")
plt.ylabel("Tần suất")
plt.show()

# Biểu đồ phân phối nhiệt độ cơ thể
plt.figure(figsize=(8, 4))
sns.histplot(df["Body_Temp"], kde=True, color="salmon", bins=30)
plt.title("Phân phối nhiệt độ cơ thể")
plt.xlabel("Nhiệt độ cơ thể (°C)")
plt.ylabel("Tần suất")
plt.show()
```

Mối quan hệ giữa biến Calories và các biến liên quan

```python
# Mối quan hệ giữa Calories và Age
plt.figure(figsize=(8, 5))
sns.scatterplot(x="Age", y="Calories", hue="Gender", data=df, alpha=0.7)
plt.title("Calories tiêu thụ theo độ tuổi")
plt.xlabel("Age (tuổi)")
plt.ylabel("Calories")
plt.legend()
plt.show()

# Mối quan hệ giữa Calories và Height
plt.figure(figsize=(8, 5))
sns.scatterplot(x="Height", y="Calories", hue="Gender", data=df, alpha=0.7)
plt.title("Calories tiêu thụ theo chiều cao")
plt.xlabel("Height (cm)")
plt.ylabel("Calories")
plt.legend()
plt.show()

# Mối quan hệ giữa Calories và Weight
plt.figure(figsize=(8, 5))
sns.scatterplot(x="Weight", y="Calories", hue="Gender", data=df, alpha=0.7)
plt.title("Calories tiêu thụ theo cân nặng")
plt.xlabel("Weight (kg)")
plt.ylabel("Calories")
plt.legend()
plt.show()

# Mối quan hệ giữa Calories và Duration
plt.figure(figsize=(8, 5))
sns.scatterplot(x="Duration", y="Calories", hue="Gender", data=df, alpha=0.7)
plt.title("Calories tiêu thụ theo thời gian tập luyện")
plt.xlabel("Duration (phút)")
plt.ylabel("Calories")
plt.legend()
plt.show()

# Mối quan hệ giữa Calories và Heart Rate
plt.figure(figsize=(8, 5))
sns.scatterplot(x="Heart_Rate", y="Calories", hue="Gender", data=df, alpha=0.7)
plt.title("Calories tiêu thụ theo nhịp tim")
plt.xlabel("Heart Rate (bpm)")
plt.ylabel("Calories")
plt.legend()
plt.show()

# Mối quan hệ giữa Calories và Body Temperature
plt.figure(figsize=(8, 5))
sns.scatterplot(x="Body_Temp", y="Calories", hue="Gender", data=df, alpha=0.7)
plt.title("Calories tiêu thụ theo nhiệt độ cơ thể")
plt.xlabel("Body Temperature (phút)")
plt.ylabel("Calories")
plt.legend()
plt.show()
```

Ma trận tương quan giữa các đặc trưng

```python
# Loại bỏ cột User_ID
df_corr = df.drop(columns=["User_ID"])

# Vẽ ma trận tương quan giữa các đặc trưng số
plt.figure(figsize=(10, 8))
correlation = df_corr.select_dtypes(include=['float64', 'int64']).corr()
sns.heatmap(correlation, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Ma trận tương quan (đã loại bỏ User_ID)")
plt.show()
```
